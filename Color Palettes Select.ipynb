```python
import matplotlib.pyplot as plt
import numpy as np


Yellows = [('Beige', '#ffe9a9'), ('Yellow', '#fffb00')]
Blues = [('Blue', '#0433ff'), ('Blue, Light', '#3dd3f8'), ('Blue, Medium', '#5594e8'), ('Blue, Navy', '#0225bd')]
Browns = [('Brown, Dark', '#945200'), ('Brown, Light', '#ca993a'), ('Brown, Very Dark', '#66461b')]
Grays = [('Gray, Dark', '#858585'), ('Gray, Light', '#b6b6b6'), ('Pearl Silver', '#cdc7bb')]
Greens = [('Green', '#01b101'),('Green, Dark', '#0b8015')]
Oranges = [('Nectarine', '#ffb973'), ('Orange', '#ff9300')]
Reds = [('Red', '#ff2600'), ('Red, Dark', '#a00205')]
Purples = [('Purple', '#942192'), ('Pink', '#ff40ff')]

Colors_Palettes = [Yellows, Blues, Browns, Grays, Greens, Oranges, Reds, Purples]

def generatePalette(HexCode):
    R = int(HexCode[1:3], 16)
    G = int(HexCode[3:5], 16)
    B = int(HexCode[5:], 16)
    h = 100
    w = 100
    temp = np.zeros((h,w,3), np.uint8)
    temp[:, :, 0] = R
    temp[:, :, 1] = G
    temp[:, :, 2] = B
    return temp
```

```python
for colors in Colors_Palettes:
    nums = len(colors)
    f = plt.figure(num=None, figsize=(2*nums, 2*nums), dpi=80, facecolor='w', edgecolor='k')
    for i, c in enumerate(colors):
        
        f.add_subplot(1,nums, i+1)
        img = generatePalette(c[1])
        plt.title(c[0])
        plt.imshow(img)
```
